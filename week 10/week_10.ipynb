```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = 2 * x + 1
```

```python
print("X values: ", x)
print("Corresponding Y values: ", y)
```

```python
plt.plot(x, y, '--r')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegression?
```

```python
model = LinearRegression()
print(model)
```

```python
X = x[:, np.newaxis]
print(X)
print(y)
```

```python
model.fit(X, y)
```

```python
print(dir(model))
```

```python
print(model.coef_)  # slope
print(model.intercept_)  # intercept
```

```python
x = np.arange(50)
y = 2 * x + 1 + np.random.randn(50)
```

```python
plt.plot(x, y, '.r')
```

```python
X = x[:, np.newaxis]
model.fit(X, y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
theta = np.array([*model.coef_, model.intercept_]).reshape(2, 1)
x_test = np.array([0, 50]).reshape(2, 1)
x_test_as_array = np.c_[x_test, np.ones((2, 1))]
y_est = x_test_as_array.dot(theta)
print(y_est)
```

```python
plt.plot(x_test, y_est, 'b-', label="learned line of fit")
plt.plot(X, y, 'r+', label="training data")
plt.xlabel("$input(x)$", fontsize=12)
plt.ylabel("$output(y)$", rotation=90, fontsize=12)
plt.title("Fitting data into a linear regression model")
plt.legend()
```
