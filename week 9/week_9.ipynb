```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([52, 31, 32, 29, 30, 27, 27])
```

## Line Plot

```python
plt.plot(x,y,color='green', linestyle = '-.')

plt.xlabel("Weather forecast of Kathmandu for the first week of June.")
plt.ylabel("Temperature in degree celsius.")
plt.title("Weather")
```

```python
plt.plot(x, y, '--r')  # Using format strings to specify arguments
plt.show()
plt.plot(x, y, '--k')
```

```python
plt.plot(x, y, color='green', marker='+', linestyle='-')
plt.show()
```

```python
plt.plot(x, y, color='#1FEC0E', marker='D', linestyle='')
plt.show()
```

```python
plt.plot(x, y, color='#888888', marker='D', linestyle='-', markersize=9)
plt.show()
```

```python
plt.plot(x, y, color='green', alpha=1)
plt.show()
```

```python
plt.plot(x, y, color='green', alpha=0.3)
plt.plot(x, y - 3, color='green', alpha=1)
plt.show()
```

## Axes labels, Legend, and Grid


```python
days = [1,2,3,4,5,6,7]
max_temp = [35,37,34,35,36,37,32]
min_temp = [30,29,29,28,22,19,22]
avg_temp = [30,32,29,29,32,31,25]
```

```python
plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
plt.show()
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
plt.show()
```

```python
from cProfile import label

from pyparsing import alphanums


plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_temp, label="Max")
plt.plot(days, min_temp, label= "Min")
plt.plot(days, avg_temp, label="Average")
plt.legend(loc = 'best', shadow=False, fontsize='large')
plt.grid(alpha=0.3)
plt.show()
```

## Bar Plot

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'SNAP']
revenue = [90, 136, 89, 27]
profit = [40, 27, 34, 3]
```

```python
xpos = np.arange(len(company))
print(xpos)
```

```python
plt.bar(xpos, revenue)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.bar(xpos, revenue)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue(bn)")
plt.title("Top S&P 500 Revenues")
plt.bar(xpos, revenue, label="Revenue")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue(bn)")
plt.title("Top S&P 500 Revenues")
plt.bar(xpos, revenue, label="Revenue")
plt.bar(xpos, profit, label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue(bn)")
plt.title("Top S&P 500 Revenues")
plt.bar(xpos - 0.2, revenue, label="Revenue")
plt.bar(xpos + 0.2, profit, label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue(bn)")
plt.title("Top S&P 500 Revenues")
plt.bar(xpos - 0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos + 0.2, profit, width=0.4, label="Profit")
plt.legend()
plt.show()
```

```python
ypos = np.arange(len(company))

plt.yticks(ypos, company)
plt.xlabel("Revenue(bn)")
plt.title("Top S&P 500 Revenues")
plt.barh(xpos - 0.2, revenue, 0.4, label="Revenue")
plt.barh(xpos + 0.2, profit, 0.4, label="Profit")
plt.legend()
plt.show()
```

## Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8)  # by default number of bins is set to 10
plt.show()
```

```python
plt.hist(blood_sugar, rwidth=0.5, bins=15)
plt.show()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.8, color='g')
plt.show()
```

```python
# Mutiple data samples in a histogram

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'])
plt.legend()
plt.show()
```

```python
# histtype=step
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, histtype='step')
plt.show()
```

```python
expense_vals = [1400, 600, 300, 410, 250]
expense_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
plt.pie(expense_vals, labels=expense_labels)
plt.show()
```

## Pie Chart


```python
plt.pie(expense_vals, labels=expense_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(expense_vals, labels=expense_labels, shadow=True, autopct='%1.1f%%', radius=1.5)
plt.show()
```

## Explode

```python
plt.axis("equal")
plt.pie(expense_vals, labels=expense_labels, shadow=True, autopct='%1.1f%%', radius=1.5, explode=[0, 0, 0, 0.1, 0.2])
plt.show()
```

## counterclock and angle properties

```python
plt.axis("equal")
plt.pie(expense_vals, labels=expense_labels, shadow=True, autopct='%1.1f%%', radius=1.5, explode=[0, 0, 0, 0.1, 0.2], counterclock=False, startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(expense_vals, labels=expense_labels, shadow=True, autopct='%1.1f%%', radius=1.5, explode=[0, 0, 0, 0.1, 0.2], counterclock=True, startangle=45)
plt.savefig("Test.png")
```

```python
plt.axis("equal")
plt.pie(expense_vals, labels=expense_labels, radius=2, autopct='%0.1f%%', explode=[0, 0.1, 0.1, 0, 0])
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=1, transparent=True)
```
